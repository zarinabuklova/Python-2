```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
def movies(row):
    if row['rating'] <= 2.0:
        return 'низкий рейтинг'
    if row['rating'] >= 2.0 and row['rating'] <= 4.0:
        return 'средний рейтинг'
    return  'высокий рейтинг'
```

```python
ratings['class'] = ratings.apply(movies, axis = 1)
```

```python
import pandas as pd
```

```python
keywords = pd.read_csv('keywords.csv')
keywords.head(150)
```

```python
def classificator(row):
    if ('москв' in row['keyword']) or ('тул' in row['keyword']) or ('ярославл' in row['keyword']): 
        return 'Центр'
    if ('петербург' in row['keyword']) or ('псков' in row['keyword']) or ('мурманск' in row['keyword']) : 
        return 'Северо-Запад'
    if ('владивосток'in row['keyword']) or ('сахалин' in row['keyword']) or ('хабаровск' in row['keyword']) : 
        return 'Дальний Восток '
    return 'undefined'
```

```python
keywords['region'] = keywords.apply(classificator, axis = 1)
```
